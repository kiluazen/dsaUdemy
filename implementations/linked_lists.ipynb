```python
class SinglyLinkedList(Node):
    def __init__(self):
        self.head = None; self.tail = None
        self.length = 0
    def push(self, val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
            self.tail= self.head
        else:
            self.tail.next = newNode
            self.tail = newNode
        self.length +=1
        return
    def pop(self):
        var = self.head
        if var is None:
            print("Undefined")
        while var:
            if var.next.next is None:
                self.tail = var
                var.next = None
                self.length -=1
                return
            var = var.next
        del var
        return
    def shift(self):
        if self.head is None:
            return KeyError
        var = self.head
        self.head = var.next
        del var
        self.length-=1
        return
    def unshift(self,value):
        var = self.head
        self.head = Node(value)
        self.head.next = var
        self.length+=1
        return
    def printlist(self):
        var = self.head
        while var:
            print(var.val)
            var = var.next
    def get(self, index):
        if index<0 or index>= self.length:
            return None
        count = 0
        result = self.head
        while(count != index):
            result = result.next
            count+=1
        return result
    def set(self, index, val):
        node = self.get(index)
        if node is not None:
            node.val = val
            return True
        return False
    def insert(self, index, val):
        if index<0 or index> self.length:
            return False
        if index == self.length:
            return self.push(val)
        if index == 0:
            return self.unshift(val)
        newNode  = Node(val)
        prev = self.get(index -1)
        current = prev.next
        prev.next = newNode
        newNode.next = current
        self.length +=1
        return True
    def remove(self, index):
        if index<0 or index>= self.length:
            return None
        if index == 0:
            return self.shift()
        if index == self.length -1:
            return self.pop()
        prev_node = self.get(index-1)
        removed = prev_node.next
        prev_node.next = removed.next
        self.length -=1
        return removed
    def reverse(self):
        node = self.head
        self.head = self.tail
        self.tail = node
        prev = None

        for i in range(self.length):
            next = node.next
            node.next = prev
            prev = node
            node = next

               

list = SinglyLinkedList()
print('-------PUSH---------')
list.push('Name:')
list.push('Kushal')
list.push('Sokke')
list.push('Malleshappa')
list.push('fuck you!')

print(list.length)
list.printlist()
print(f'Head of the list = {list.head}')

list.pop()
print('-------POP---------')
print(list.length)
list.printlist()

list.shift()
print('---------SHIFT-------')
print(list.length)
list.printlist()

list.unshift('Co-Fouder')
print('------UNSHIFT----------')
print(list.length)
list.printlist()
print('-------GET---------')
print(list.get(2))

print('-------SET---------')
list.set(2,'Gejjagundanalli')
list.printlist()
print('------INSERT---------')
list.insert(2,'Asshole')
list.printlist()
print('------REMOVE---------')
list.remove(3)
list.printlist()
print('------REVERSE---------')
list.reverse()
list.printlist()
```
