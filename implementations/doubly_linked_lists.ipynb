```python
class Node():
    def __init__(self, val) -> None:
        self.val = val
        self.next = None
        self.prev= None
    def __str__(self) -> str:
        return self.val
```

```python
x = Node('Riddhima')
x.next = 'Kushal'
x.prev = 'Aah! Oh I love it and I hate it at the same time'
print(x)
print(x.prev)
```

```python
class DoubleLikedList(Node):
    def __init__(self) -> None:
        self.head = None
        self.tail = None
        self.length = 0
    def printlist(self):
        var = self.head
        while var:
            print(var.val)
            var = var.next
    def push(self,val):
        newNode = Node(val)
        if self.length == 0:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode
        self.length +=1
        return 
    def pop(self):
        if self.length == 0:
            return None
        pop_node = self.tail
        if self.length ==1:
            self.head = None
            self.tail = None
        self.tail = pop_node.prev
        self.tail.next = None
        pop_node.prev = None
        self.length -=1
        return
    def shift(self):
        if self.length ==0:
            return None
        if self.length ==1:
            self.head = None
            self.tail = None
        old_head = self.head
        self.head = old_head.next
        self.head.prev = None
        old_head.next = None
        self.length -=1
        return
    def get(self,index):
        if index<0 or index>=self.length:
            return None
        if index < self.length /2:
            count = 0; current = self.head
            while count!= index:
                current = current.next
                count+=1
        else :
            count = self.length -1; current = self.tail
            while count!= index:
                current = current.prev
                count -=1 

        return current
    def set(self, index, val):
        node = self.get(index)
        if node is not None:
            node.val = val
            return True
        return False
    
    def insert( self, index, val):
        new_node = Node(val)
        if index<0 or index>self.length:
            return False
        if index == self.length:
            self.push(val)
            return True
        prev_node = self.get(index-1)
        next_node = self.get(index)
        prev_node.next = new_node
        new_node.prev = prev_node
        new_node.next = next_node
        next_node.prev = new_node
        self.length+=1
        return True
    def remove(self, index):
        if index<0 or index>=self.length:
            return False
        if index == 0:
            return self.shift() 
        if index == self.length -1 :
            return self.pop()
        removed_node = self.get(index)
        before_node = removed_node.prev
        after_node = removed_node.next
        before_node.next = after_node
        after_node.prev = before_node
        removed_node.next = None; removed_node.prev = None
        self.length -=1
        return True

list = DoubleLikedList()
list.push('damn')
list.push('Kushal')
list.push('Riddhima')
list.push('Jennifer')
list.push('Emma Stone')
list.push('Geharaiyaan')
list.push('Timothee')
list.push('lol')


list.printlist()
oldtail = list.tail
print("-------- ---POP--- --------")
list.pop()
list.printlist()
print("-------- ---SHIFT--- --------")
list.shift()
list.printlist()
print("-------- ---GET--- --------")
print(list.get(2))
print("-------- ---SET---- --------")
list.set(1, 'Crush')
list.printlist()
print("-------- ---INSERT---- --------")
list.insert(4, 'No Hard Feelings')
list.printlist()
print("-------- ---REMOVE---- --------")
list.remove(4)
list.printlist()
print("-------- ------- --------")
print(list.get(5))
```
